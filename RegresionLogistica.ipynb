#Precision:
Este indica la relación entre la predicciones correctas y las predicciones correctas esperadas, en este caso podemos ver como arroja un valor alto, por lo que se puede interpretar que concuerda el modelo entrenado con los datos que se le indico, mostrando un alto grado de concordancia

#Recall:
Se usa para saber cuantos valores positivos son correctamente clasificados como podemos ver el valor supera el 90% Tambien se puede interpretar que más 90% de los valores clasificados son verdaderos positivos

#f1 score:
Esta combina la precisión y el recall, para obtener un valor mucho más objetivo, entre más alto su valor se interpreta como mejor modelo. De nuevo ya que supera más del 90%, el modelo hizo una predicción correcta en todo el conjunto de datos
